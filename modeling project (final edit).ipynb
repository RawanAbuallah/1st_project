## <span style='color:#AC327E'> 1.Data Table </span> 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {'x_cm':range(20,100+1,20),"t1_msec":[316.99,543.32,737.89,846.91,1150],
        "t2_msec":[333.43,562.08,688.17,830.18,1140]}
Data_table = pd.DataFrame(data)
```

```python
Data_table
```

## <span style='color:#AC327E'> 2.Calculation of Data Table </span> 

```python
t = lambda t1 , t2 : ((t1+t2)/2)
Data_table["average_t_msec"]= t (Data_table.t1_msec , Data_table.t2_msec)
```

```python
Data_table
```

```python
t = lambda t1 : t1/1000
Data_table["average_t_sec"]= t(Data_table["average_t_msec"])
```

```python
Data_table
```

```python
t = lambda t1 : t1**2
Data_table["average_t^2_sec^2"] = t(Data_table["average_t_sec"])
```

```python
Data_table
```

```python
Data_table.rename(columns={"x_cm":"x(cm)"},inplace=True)
Data_table.rename(columns={"t1_msec":"t1(msec)"},inplace=True)
Data_table.rename(columns={"t2_msec":"t2(msec)"},inplace=True)
Data_table.rename(columns={"average_t_msec":"average_t(msec)"},inplace=True)
Data_table.rename(columns={"average_t_sec":"average_t(sec)"},inplace=True)
Data_table.rename(columns={"average_t^2_sec^2":"average_t^2(sec^2)"},inplace=True)
def color(val):
    color = '#4E7C97' 
    return 'color: %s' % color
Data_table.style.applymap(color)
```

## <span style='color:#AC327E'> 3.Editing Table </span> 

```python
import seaborn as sns

cm = sns.light_palette("pink", as_cmap=True)

s = Data_table.style.background_gradient(cmap=cm)
s
```

## <span style='color:#AC327E'> 4.Plotting Table </span> 

```python
import seaborn as sns
x = Data_table["average_t^2(sec^2)"]
y = Data_table['x(cm)']
sns.set()
# draw the fitting line to find the slope. 
plt.plot(x,y,"o",color='#612D55')
# calculate the slope and y-intercept.
# Acoording to our equation , it is linear equation so it is the first order ( so the digit will be 1 ).
# furthermore , if you want to draw second order or third order , you can but 2 , 3 respectively .

m, b = np.polyfit(x, y, 1)
# the fitting line.
plt.plot(x, m*x + b,color='#58A6B0')
plt.title("Relationship between x(cm) VS t^2(sec^2)",fontsize=30,font="Aharoni",color='#013460')
plt.xlabel("t^2(sec^2)",fontsize=30,font="Aharoni",color='#942776')
plt.ylabel("x(cm)",fontsize=30,font="Aharoni",color='#942776')

plt.show()
```

## <span style='color:#AC327E'> 7.Theoretical accelration a_th </span>
### <span style='color:#B48FAF'> m--> (Hanging Masses) , M--> (Cart Mass) </span> 

```python
m=50
M=207.07
g=980

a_th= ((m)/(m+M))*g
print("theoretical accelration , a_th=",round(a_th,2),"cm/s^2")
```

## <span style='color:#AC327E'> 8.Experimental accelration a_exp from python's slope </span>

```python
a_exp1= 2*m
print("experimental accelration , a_exp=",round(a_exp1,2),"cm/s^2")
```

## <span style='color:#AC327E'> 9.Exrerimental error for python's slope</span>

```python
import math as m

diff1= m.fabs((a_exp1-a_th)/(a_th))*100
print("Exrerimental error=",diff1 ,"%")
```

## <span style='color:#AC327E'> 10.Slope from selected points </span>

```python
# edit *******
# points  , P1=(0.3,40) , P2=(0.4,50)
x1 , y1 = 0.3 , 40
x2 , y2 = 0.4 , 50
slope= (y2-y1)/(x2-x1)
slope
```

## <span style='color:#AC327E'> 11.Experimental accelration a_exp from selected points </span>

```python
a_exp= 2*slope
print("experimental accelration , a_exp=",round(a_exp,2),"cm/s^2")
```

## <span style='color:#AC327E'> 12.Exrerimental error from selected points </span>

```python
import math as m

diff= m.fabs((a_exp-a_th)/(a_th))*100
print("Exrerimental error=",diff ,"%")
```
